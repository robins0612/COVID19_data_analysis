```python
#importing librarires
import pandas as pd
import plotly.express as px
```

```python
#importing dataset
df = pd.read_csv('country_wise_latest.csv')
```

```python
#checking the dimension
df.shape
```

```python
#displaying the data
df.head()
```

```python
#checking the null values
df.isnull().sum()
```

## EDA (Exploratory Data Analysis) and Visualization

### 1. Which countries had the highest number of confirmed cases?

```python
hight_confirmed_cases = df.nlargest(10,'Confirmed')
```

```python
px.bar(hight_confirmed_cases,x='Country/Region',y='Confirmed',title='Countries with highest number of confirmed cases')
```

### 2. What was the distribution of confirmed cases across different WHO regions?

```python
df['WHO Region'].unique()
```

```python
px.pie(df,names='WHO Region',values='Confirmed',title='Distribution of confirmed cases across different WHO regions')
```

### 3. What were the trends in new cases, new deaths, and new recoveries in each country?

```python
px.scatter(df,x='New cases',y='New deaths',size = 'New recovered', color = 'Country/Region', title='Trends in new cases in each country')
```

### 4. What was the 1-week percentage increase in confirmed cases for each country?


```python
px.choropleth(df,
              locations='Country/Region',
              locationmode='country names',
              color='1 week % increase',
              hover_name='Country/Region',
              title='1-week percentage increase in confirmed cases for each country',
              color_continuous_scale= px.colors.sequential.Turbo_r)
```

### 5. How do death rates per 100 confirmed cases vary by WHO region?

```python
px.box(df,x='WHO Region',y='Deaths / 100 Cases',title='Death rates per 100 confirmed cases vary by WHO region')
```

### 6. What was the mortality rate (deaths per 100 confirmed cases) for each country?

```python
 px.choropleth(df,
              locations='Country/Region',
              locationmode='country names',
              color='Deaths / 100 Cases',
              hover_name='Country/Region',
              title='Mortality rate (deaths per 100 confirmed cases) for each country',
              color_continuous_scale= px.colors.sequential.R)
```

### 7. How many countries are in each WHO *region*?

```python
who_region_count = df['WHO Region'].value_counts().reset_index()
```

```python
px.bar(who_region_count,x='WHO Region',y='count',title='No of Countries in each WHO region')
```

### 8. Treemap of COVID-19 confirmed cases by WHO region and country?

```python
px.treemap(df,path=['WHO Region','Country/Region'],values='Confirmed',title='Treemap of COVID-19 confirmed cases by WHO region and country')
```
